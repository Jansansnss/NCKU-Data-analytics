    決策論中 （如風險管理），決策樹（Decision tree）由一個決策圖和可能的結果（包括資源成本和風險）組成， 用來創建到達目標的規劃。決策樹建立並用來輔助決策，是一種特殊的樹結構。決策樹是一個利用像樹一樣的圖形或決策模型的決策支持工具，包括隨機事件結果，資源代價和實用性。它是一個算法顯示的方法。決策樹經常在運籌學中使用，特別是在決策分析中，它幫助確定一個能最可能達到目標的策略。如果在實際中，決策不得不在沒有完備知識的情況下被在線採用，一個決策樹應該平行概率模型作為最佳的選擇模型或在線選擇模型算法。決策樹的另一個使用是作為計算條件概率的描述性手段。

14. **使用 ensemble 進行實驗|+8|**
    
    AdaBoost，是英文"Adaptive Boosting"（自適應增強）的縮寫，是一種機器學習方法。AdaBoost方法的自適應在於：前一個分類器分錯的樣本會被用來訓練下一個分類器。AdaBoost方法中使用的分類器可能很弱（比如出現很大錯誤率），但只要它的分類效果比隨機好一點（比如兩類問題分類錯誤率略小於0.5），就能夠改善最終得到的模型。而錯誤率高於隨機分類器的弱分類器也是有用的，因為在最終得到的多個分類器的線性組合中，可以給它們賦予負係數，同樣也能提升分類效果。[資料來源](https://zh.wikipedia.org/zh-tw/AdaBoost)
    
    在集成算法中，bagging方法形成了一類算法，它在原始訓練集的隨機子集上建立一個黑箱估計的多個實例，然後将它們各自的預測聚合起來形成最终的預測。這些方法通過在基本估計量（如決策樹）的構造過程中引入隨機化，然後對其進行集成，來降低其方差。在許多情况下，bagging方法是相對於單個模型進行改進的一種非常簡單的方法，而無需調整底層的基本算法。由於bagging方法提供了一種減少過度撮合的方法，因此它最適合於强模型和複雜模型（例如，充分開發的決策樹），而boosting方法通常最適合於弱模型（例如，淺層決策樹）。[資料來源](https://blog.csdn.net/weixin_46072771/article/details/106319323)

## 模型訓練

請在此區塊後使用**一個或多個區塊**進行模型訓練，**需要**包含以下內容：

- **選擇模型**
    - 設定模型**超參數**
    - **控制隨機亂數**
- 計算**準確度（Accuracy）**
    - 使用 **5-fold cross validation**，輸出**平均準確度**
    - 最後再使用**所有**資料進行訓練並輸出**準確度**
    - 準確度必須**超越**課堂使用的範例
    
如果有進行多於一個以上的實驗，**可以保留所有**的實驗結果，但是執行時間將會納入**全部**實驗。

# Overfitting

````python
Overfitting_train_x = df[['Sex', 'Age', 'Pclass', 'SibSp', 'Parch', 'Embarked', 'Fare']].copy()         # 取出訓練資料需要分析的資料欄位
Overfitting_train_y = df['Survived'].copy()                 # 取出訓練資料的答案

from sklearn.impute import SimpleImputer       # 匯入填補缺失值的工具
from sklearn.preprocessing import LabelEncoder # 匯入 Label Encoder

le = LabelEncoder()                            # 創造 Label Encoder
le.fit(Overfitting_train_x['Sex'])                         # 給予每個類別一個數值
Overfitting_train_x['Sex'] = le.transform(Overfitting_train_x['Sex'])  # 轉換所有類別成為數值

lf = LabelEncoder()                            # 創造 Label Encoder
lf.fit(Overfitting_train_x['Embarked'])                         # 給予每個類別一個數值
Overfitting_train_x['Embarked'] = lf.transform(Overfitting_train_x['Embarked'])  # 轉換所有類別成為數值

Overfitting_train_x = Overfitting_train_x.fillna(-1)

# 模型訓練

from sklearn.model_selection import KFold             # 匯入 K 次交叉驗證工具
from sklearn.tree import DecisionTreeClassifier       # 匯入決策樹模型
from sklearn.metrics import accuracy_score            # 匯入準確度計算工具

kf = KFold(n_splits=5,                                # 設定 K 值
           random_state=9487,
           shuffle=True)
kf.get_n_splits(Overfitting_train_x)                              # 給予資料範圍

Overfitting_train_acc_list = []                                   # 儲存每次訓練模型的準確度
valid_acc_list = []                                   # 儲存每次驗證模型的準確度

for Overfitting_train_index, valid_index in kf.split(Overfitting_train_x):    # 每個迴圈都會產生不同部份的資料
    Overfitting_train_x_split = Overfitting_train_x.iloc[Overfitting_train_index]         # 產生訓練資料
    Overfitting_train_y_split = Overfitting_train_y.iloc[Overfitting_train_index]         # 產生訓練資料標籤
    valid_x_split = Overfitting_train_x.iloc[valid_index]         # 產生驗證資料
    valid_y_split = Overfitting_train_y.iloc[valid_index]         # 產生驗證資料標籤
    
    model = DecisionTreeClassifier(random_state=1012) # 創造決策樹模型
    model.fit(Overfitting_train_x_split, Overfitting_train_y_split)           # 訓練決策樹模型
    
    Overfitting_train_pred_y = model.predict(Overfitting_train_x_split)       # 確認模型是否訓練成功
    Overfitting_train_acc = accuracy_score(Overfitting_train_y_split,         # 計算訓練資料準確度
                               Overfitting_train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       # 驗證模型是否訓練成功
    valid_acc = accuracy_score(valid_y_split,         # 計算驗證資料準確度
                               valid_pred_y)
    
    Overfitting_train_acc_list.append(Overfitting_train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    '    min train accuracy: {}\n' +
    '    max train accuracy: {}\n' +
    'average valid accuracy: {}\n' +
    '    min valid accuracy: {}\n' +
    '    max valid accuracy: {}').format(
    np.mean(Overfitting_train_acc_list),                          # 輸出平均訓練準確度
    np.min(Overfitting_train_acc_list),                           # 輸出最低訓練準確度
    np.max(Overfitting_train_acc_list),                           # 輸出最高訓練準確度
    np.mean(valid_acc_list),                          # 輸出平均驗證準確度
    np.min(valid_acc_list),                           # 輸出最低驗證準確度
    np.max(valid_acc_list)                            # 輸出最高驗證準確度
))
````
